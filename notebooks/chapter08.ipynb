## 問題 8-1: Cat の継承と Sleep

Cat は Animal のプロパティ（Name・Age）と ShowProfile を引き継いでいます。

```vb.net
Dim cat As New Cat("タマ", 2)
Console.WriteLine(cat.ShowProfile())
Console.WriteLine(cat.Sleep())
```

## 問題 8-2: Dog の継承と Run

```vb.net
Dim dog As New Dog("ポチ", 3)
Console.WriteLine(dog.ShowProfile())
Console.WriteLine(dog.Run())
```

## 問題 8-3 / 8-4: Speak のオーバーライド

Animal の Speak は "......" を返しますが、Cat と Dog はそれぞれ上書きしています。

```vb.net
Dim cat As New Cat("タマ", 2)
Dim dog As New Dog("ポチ", 3)
Console.WriteLine($"Cat.Speak: {cat.Speak()}")
Console.WriteLine($"Dog.Speak: {dog.Speak()}")
```

## 問題 8-5: ポリモーフィズム

Animal 型の配列に Cat と Dog を混在させ、同じ Speak() 呼び出しで異なる結果が出ることを確認します。

```vb.net
Dim animals As Animal() = {
    New Cat("タマ", 2),
    New Dog("ポチ", 3),
    New Cat("ミケ", 1),
    New Dog("コロ", 5)
}

For Each a In animals
    Console.WriteLine($"{a.ShowProfile()} → {a.Speak()}")
Next
```

```vb.net
' Animal 型変数に Cat インスタンスを代入してみる
Dim a As Animal = New Cat("サクラ", 4)
Console.WriteLine($"変数の型: Animal、実際のインスタンス: Cat")
Console.WriteLine($"Speak の結果: {a.Speak()}")   ' Cat.Speak が呼ばれる
```
