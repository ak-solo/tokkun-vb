## 問題 6-7: 素数判定

任意の数が素数かどうか確認してみましょう。

```vb.net
Dim n = 97
Dim result = If(Exercises.Problem6_7(n), "素数", "素数ではない")
Console.WriteLine($"{n} は{result}")
```

```vb.net
' 1000 未満の素数をすべて表示する
Dim primes As New List(Of Integer)
For i = 2 To 999
    If Exercises.Problem6_7(i) Then primes.Add(i)
Next
Console.WriteLine($"1000 未満の素数: {primes.Count} 個")
Console.WriteLine(String.Join(", ", primes))
```
