## 問題 1-11 vs 1-12: `/` と `\` の違い

VB.NET では Integer 同士でも `/` を使うと結果が **Double** になります。  
整数の商だけ欲しい場合は `\`（バックスラッシュ）を使います。

```vb.net
Console.WriteLine($"10 / 3  = {Exercises.Problem1_11()} （Double）")
Console.WriteLine($"10 \ 3 = {Exercises.Problem1_12()} （Integer）")
```
